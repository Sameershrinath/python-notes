```python
#Code to open the camera on execution
import cv2

def main():
    # Open the default camera (usually the webcam)
    cap = cv2.VideoCapture(0)

    # Check if the camera was opened successfully
    if not cap.isOpened():
        print("Error: Unable to open camera.")
        return

    # Loop to continuously capture frames from the camera
    while True:
        # Read a frame from the camera
        ret, frame = cap.read()

        # Check if the frame was read successfully
        if not ret:
            print("Error: Unable to read frame.")
            break

        # Display the frame in a window named 'Camera'
        cv2.imshow('Camera', frame)

        # Check for the 'q' key to quit the program
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    # Release the camera and close all OpenCV windows
    cap.release()
    cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
```

```python
import cv2
import speech_recognition as sr
import tkinter as tk

def open_camera():
    # Open the default camera (usually the webcam)
    cap = cv2.VideoCapture(0)

    # Check if the camera was opened successfully
    if not cap.isOpened():
        print("Error: Unable to open camera.")
        return

    # Loop to continuously capture frames from the camera
    while True:
        # Read a frame from the camera
        ret, frame = cap.read()

        # Check if the frame was read successfully
        if not ret:
            print("Error: Unable to read frame.")
            break

        # Display the frame in a window named 'Camera'
        cv2.imshow('Camera', frame)

        # Check for the 'q' key to quit the program
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    # Release the camera and close all OpenCV windows
    cap.release()
    cv2.destroyAllWindows()

def take_picture():
    # Open the default camera (usually the webcam)
    cap = cv2.VideoCapture(0)

    # Check if the camera was opened successfully
    if not cap.isOpened():
        print("Error: Unable to open camera.")
        return

    # Capture a single frame
    ret, frame = cap.read()

    # Check if the frame was read successfully
    if not ret:
        print("Error: Unable to read frame.")
        return

    # Save the captured frame to an image file
    cv2.imwrite("captured_image.jpg", frame)
    print("Image captured successfully!")

    # Release the camera
    cap.release()

def listen_command():
    # Initialize recognizer class
    recognizer = sr.Recognizer()

    # Use the microphone as source for audio input
    with sr.Microphone() as source:
        print("Listening for command...")
        recognizer.adjust_for_ambient_noise(source)
        audio = recognizer.listen(source)

    try:
        print("Recognizing command...")
        command = recognizer.recognize_google(audio).lower()
        print("Command:", command)
        if 'open camera' in command:
            open_camera()
        elif 'take picture' in command:
            take_picture()
        else:
            print("Command not recognized.")

    except sr.UnknownValueError:
        print("Sorry, I couldn't understand the audio.")
    except sr.RequestError as e:
        print("Request error:", e)

def start_listening():
    listen_command()

def main():
    root = tk.Tk()
    root.title("Voice Command Camera")

    start_button = tk.Button(root, text="Start Listening", command=start_listening)
    start_button.pack()

    root.mainloop()

if __name__ == "__main__":
    main()
```

```python
import cv2
import os
import speech_recognition as sr
import pyttsx3
from PIL import Image

# Function to capture a picture
def capture_picture():
    camera = cv2.VideoCapture(0)
    return_value, image = camera.read()
    cv2.imwrite('captured_image.jpg', image)
    del(camera)

# Function to recognize voice command
def recognize_voice():
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening for voice command...")
        audio = recognizer.listen(source)

    try:
        command = recognizer.recognize_google(audio)
        return command.lower()
    except sr.UnknownValueError:
        return "unknown"
    except sr.RequestError:
        return "request error"

# Function to display the captured image
def display_image(image_path):
    img = Image.open(image_path)
    img.show()

# Function to speak out the given text
def speak(text):
    engine = pyttsx3.init()
    engine.say(text)
    engine.runAndWait()

# Main function
def main():
    while True:
        command = recognize_voice()
        if command == "click pictures":
            print("Capturing picture...")
            capture_picture()
            speak("Done sir")
            display_image('captured_image.jpg')

if __name__ == "__main__":
    main()
```

```python
import tkinter as tk
from tkinter import PhotoImage, Entry, Button
import os
import speech_recognition as sr
import pyttsx3
import webbrowser
import sys

# Initialize Text-to-Speech engine
engine = pyttsx3.init()

# Function to speak text
def speak(text):
    engine.say(text)
    engine.runAndWait()

# Function to listen to user input
def listen():
    r = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening...")
        r.adjust_for_ambient_noise(source)
        audio = r.listen(source)
    
    try:
        print("Recognizing...")
        query = r.recognize_google(audio)
        print("You said:", query)
        return query
    except sr.UnknownValueError:
        print("Sorry, I didn't understand.")
        return ""
    except sr.RequestError as e:
        print("Could not request results; {0}".format(e))
        return ""

# Function to handle button click event
def on_button_click():
    speak("Hello sir")
    response = listen()

    if response.lower() == "hello":
        speak("How may I help you?")
        query = listen()

        if query:
            # Perform Google search
            search_url = "https://www.google.com/search?q=" + query.replace(" ", "+")
            webbrowser.open(search_url)
            speak("Here are the search results for " + query)

            # Ask if there's anything else the user wants
            speak("Anything else, sir?")
            next_query = listen()

            while next_query.lower() != "no":
                if next_query.lower() == "yes brother":
                    speak("What else can I search for you?")
                    query = listen()

                    if query:
                        # Perform Google search
                        search_url = "https://www.google.com/search?q=" + query.replace(" ", "+")
                        webbrowser.open(search_url)
                        speak("Here are the search results for " + query)

                        # Ask if there's anything else the user wants
                        speak("Anything else, sir?")
                        next_query = listen()
                    else:
                        speak("Sorry, I didn't catch that. Anything else, sir?")
                        next_query = listen()
                else:
                    speak("Sorry, I didn't catch that. Anything else, sir?")
                    next_query = listen()

            speak("Goodbye sir, have a nice day!")
            sys.exit()  # Exit the program after finishing the conversation

# Function to exit the program
def exit_program():
    sys.exit()

# Create main window
root = tk.Tk()
root.title("Voice Assistant")

# Make the window full screen
root.attributes('-fullscreen', True)

# Load images
images = []
for f in os.listdir('.'):
    if f.endswith('.png') or f.endswith('.jpg'):
        try:
            images.append(PhotoImage(file=f))
        except tk.TclError as e:
            print(f"Failed to load image '{f}': {e}")

# Display images
print("Number of images found:", len(images))  # Debugging statement
image_index = 0
if images:  # Check if images are loaded
    image_label = tk.Label(root, image=images[image_index])
    image_label.pack(side=tk.TOP, fill=tk.BOTH, expand=tk.YES)
else:
    print("No images found!")

# Function to change image
def change_image():
    global image_index
    image_index = (image_index + 1) % len(images)
    image_label.config(image=images[image_index])

# Create search box
search_entry = Entry(root, font=('Arial', 14))
search_entry.pack(side=tk.BOTTOM, fill=tk.X)

# Create search button
search_button = Button(root, text="Search", font=('Arial', 14), command=lambda: webbrowser.open("https://www.google.com/search?q=" + search_entry.get()))
search_button.pack(side=tk.BOTTOM, pady=10)

# Create exit button
exit_button = Button(root, text="Exit", font=('Arial', 14), command=exit_program)
exit_button.pack(side=tk.BOTTOM, pady=10)

# Create voice assistant button
voice_button = Button(root, text="Start Voice Assistant", font=('Arial', 14), bg="blue", fg="white", command=on_button_click)
voice_button.pack(side=tk.BOTTOM, pady=10)

# Run the GUI
root.mainloop()
```

```python
#this is the pretrained face detector 
import cv2

# Load the pre-trained face detector
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Set up video capture
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break
    
    # Convert frame to grayscale for faster processing
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # Detect faces in the grayscale frame
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
    
    # Draw rectangles around the detected faces
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
    
    # Display the resulting frame
    cv2.imshow('Face Detection', frame)
    
    # Exit if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the capture
cap.release()
cv2.destroyAllWindows()
```
