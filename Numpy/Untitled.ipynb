```python
class Solution:
    def findMinCost(self, x, y, costX, costY):
        m, n = len(x), len(y)
        
        # Initialize dp table
        dp = [[0] * (n + 1) for _ in range(m + 1)]
        
        # Base cases
        for i in range(1, m + 1):
            dp[i][0] = i * costX
        for j in range(1, n + 1):
            dp[0][j] = j * costY
        
        # Fill the dp table
        for i in range(1, m + 1):
            for j in range(1, n + 1):
                if x[i-1] == y[j-1]:
                    dp[i][j] = dp[i-1][j-1]
                else:
                    dp[i][j] = min(dp[i-1][j] + costX,  # Delete from x
                                   dp[i][j-1] + costY)  # Delete from y
        
        return dp[m][n]
```

```python
def sorting(arr):
    n=len(arr)
    for i in range(n):
        for j in range(0,n-i-1):
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
    return arr
```
